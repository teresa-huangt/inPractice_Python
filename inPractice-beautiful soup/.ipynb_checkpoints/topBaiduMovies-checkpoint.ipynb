# Notebook of Top Movies in Baidu.com

```python
import pandas as pd

from bs4 import BeautifulSoup

import urllib
```

## Parameters

```python
url = 'https://top.baidu.com/board?tab=movie'
```

## Fetch Content

```python
req = urllib.request.Request(url=url)

with urllib.request.urlopen(req) as f:
    content = f.read().decode('utf-8')

print(content, file=open('a.html', 'w'))
print(content)
```

## Soup it

```python
soup = BeautifulSoup(content)
soup
```

## Parse the entries

```python
def parse_entry(ent):
    ''' Parse the informations from the entry: [ent],
        a dict with name, type, actors, hot_index and desp information of the movie will be produced and returned.
    '''
    
    hot_index = int(ent.find('div', {'class': 'hot-index_1Bl1a'}).text)
    name = ent.find('div', {'class': 'c-single-text-ellipsis'}).text
    stuff = ent.find_all('div', {'class': 'intro_1l0wp'})
    desp = ent.find('div', {'class': 'c-single-text-ellipsis desc_3CTjT'}).text
    
    return dict(
        name=name,
        type=stuff[0].text,
        actors=stuff[1].text,
        hot_index=hot_index,
        desp=desp
    )
```

```python
raw_entries = soup.find_all('div', {'class': 'category-wrap_iQLoo'})
print(f'Found raw_entries of {len(raw_entries)}')
```

```python
raw_entries[0]
```

```python
df = pd.DataFrame([parse_entry(ent) for ent in raw_entries])
df
```
